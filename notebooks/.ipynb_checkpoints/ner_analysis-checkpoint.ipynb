# Understanding Cybersecurity NER Tags

This dataset uses BIO (Beginning, Inside, Outside) tagging scheme to mark cybersecurity-related entities in text. Here's what each tag represents:

## Tag Types and Examples

### 1. Indicators (B-Indicator, I-Indicator)
- Marks technical indicators of cyber threats
- Examples:
  - IP addresses and ports: "185.165.28.15:7878"
  - Package names: "com.fg.wallet"
  - Network paths: "7878/tcp"

### 2. Systems (B-System, I-System)
- Identifies software systems, platforms, and services
- Examples:
  - Platforms: "VMWare", "Next-Generation"
  - Software components: "Services", "Market"
  - Infrastructure elements: "Bignox"

### 3. Malware (B-Malware, I-Malware)
- Tags malicious software names and types
- Examples:
  - Malware families: "FakeSpy", "SandroRAT"
  - Malware types: "spyware"
  - Variants: "DroidPlugin"

### 4. Organizations (B-Organization, I-Organization)
- Marks company and organization names
- Examples:
  - Companies: "Barclays", "Qihoo"
  - Security firms: "Lookout"
  - Divisions: "Technologies"

### 5. Vulnerabilities (B-Vulnerability, I-Vulnerability)
- Indicates security vulnerabilities and weaknesses
- Examples:
  - CVE numbers: "CVE-2015-2387"
  - Vulnerability types: "heap overflow"
  - Attack vectors: "local kernel"

### 6. Outside Tag (O)
- Marks words that aren't part of named entities
- Examples: "few", "July", "By"

## Tag Prefix Meanings
- **B-**: Marks the Beginning of an entity
- **I-**: Marks the Inside (continuation) of an entity
- **O**: Outside (not part of any entity)

## Note
One entity can span multiple words. For example:
- "Next-Generation[B-System] Threat[I-System]" is one complete system entity
- "heap[B-Vulnerability] overflow[I-Vulnerability]" is one complete vulnerability entity

````python
import pandas as pd
import numpy as np
from collections import defaultdict
import matplotlib.pyplot as plt
import seaborn as sns

def perform_basic_eda(file_path):
    """
    Perform basic Exploratory Data Analysis on cybersecurity NER dataset
    
    Parameters:
    -----------
    file_path : str
        Path to the NER-tagged data file
    """
    # Step 1: Data Loading and Basic Statistics
    print("Step 1: Basic Dataset Statistics")
    print("-" * 50)
    
    # Load data into pandas DataFrame
    words = []
    tags = []
    with open(file_path, 'r') as f:
        for line in f:
            if line.strip():
                word, tag = line.strip().split('\t')
                words.append(word)
                tags.append(tag)
    
    df = pd.DataFrame({'word': words, 'tag': tags})
    
    # Basic counts
    total_tokens = len(df)
    unique_words = df['word'].nunique()
    unique_tags = df['tag'].nunique()
    
    print(f"Total tokens: {total_tokens:,}")
    print(f"Unique words: {unique_words:,}")
    print(f"Unique tags: {unique_tags}")
    print(f"Vocabulary density: {unique_words/total_tokens*100:.2f}%")
    
    # Step 2: Tag Distribution Analysis
    print("\nStep 2: Tag Distribution Analysis")
    print("-" * 50)
    
    tag_counts = df['tag'].value_counts()
    tag_proportions = (tag_counts / len(df) * 100).round(2)
    
    print("\nTag Distribution:")
    for tag, count in tag_counts.items():
        print(f"{tag:15} {count:7,} ({tag_proportions[tag]:5.2f}%)")
    
    # Step 3: Entity Type Analysis
    print("\nStep 3: Entity Type Analysis")
    print("-" * 50)
    
    # Group B- and I- tags
    entity_types = defaultdict(int)
    for tag, count in tag_counts.items():
        if tag != 'O':
            entity_type = tag[2:]  # Remove B- or I- prefix
            entity_types[entity_type] += count
    
    print("\nEntity Type Distribution:")
    for entity_type, count in sorted(entity_types.items(), key=lambda x: x[1], reverse=True):
        print(f"{entity_type:15} {count:7,} ({count/total_tokens*100:.2f}%)")
    
    # Step 4: Word Length Analysis
    print("\nStep 4: Word Length Analysis")
    print("-" * 50)
    
    df['word_length'] = df['word'].str.len()
    
    length_stats = df.groupby('tag')['word_length'].agg(['mean', 'std', 'min', 'max'])
    print("\nWord Length Statistics by Tag:")
    print(length_stats.round(2))
    
    # Step 5: Entity Sequence Analysis
    print("\nStep 5: Entity Sequence Analysis")
    print("-" * 50)
    
    # Analyze transitions between tags
    df['next_tag'] = df['tag'].shift(-1)
    transitions = df.groupby(['tag', 'next_tag']).size()
    
    print("\nMost Common Tag Transitions:")
    print(transitions.nlargest(10))
    
    # Step 6: Statistical Tests
    print("\nStep 6: Statistical Analysis")
    print("-" * 50)
    
    # Chi-square test for tag distribution
    from scipy.stats import chisquare
    
    # Test if entity tag distribution is uniform (excluding 'O' tag)
    entity_counts = [count for tag, count in tag_counts.items() if tag != 'O']
    chi2_stat, p_value = chisquare(entity_counts)
    
    print("\nChi-square test for entity tag distribution:")
    print(f"Chi-square statistic: {chi2_stat:.2f}")
    print(f"p-value: {p_value:.10f}")
    
    # Step 7: Key Findings Summary
    print("\nStep 7: Key Findings")
    print("-" * 50)
    
    print("\nKey Statistical Findings:")
    print(f"1. Dataset size: {total_tokens:,} tokens with {unique_words:,} unique words")
    print(f"2. Named entity density: {100-tag_proportions['O']:.2f}% of tokens are named entities")
    most_common_entity = sorted(entity_types.items(), key=lambda x: x[1], reverse=True)[0]
    print(f"3. Most common entity type: {most_common_entity[0]}")    
    print(f"4. Average word length: {df['word_length'].mean():.2f} characters")
    print("5. Entity distribution is highly non-uniform (p < 0.001)")
    
    return df, tag_counts, entity_types, length_stats, transitions

# Additional utility functions for visualization
def plot_tag_distribution(tag_counts):
    plt.figure(figsize=(12, 6))
    tag_counts.plot(kind='bar')
    plt.title('Distribution of NER Tags')
    plt.xlabel('Tag')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    return plt

def plot_word_length_distribution(df):
    plt.figure(figsize=(12, 6))
    sns.boxplot(x='tag', y='word_length', data=df)
    plt.title('Word Length Distribution by Tag')
    plt.xlabel('Tag')
    plt.ylabel('Word Length')
    plt.xticks(rotation=45)
    plt.tight_layout()
    return plt

# Example usage:
# df, tag_counts, entity_types, length_stats, transitions = perform_basic_eda('train.txt')
````

# Exploratory Data Analysis of Cybersecurity NER Dataset

## Overview
The dataset is a cybersecurity-focused Named Entity Recognition (NER) corpus using BIO (Beginning, Inside, Outside) tagging scheme. This analysis reveals key patterns and characteristics that would be valuable for developing robust cybersecurity classification systems.

## Key Findings

### 1. Dataset Scale and Composition
- Total corpus size: 68,191 tokens
- Unique vocabulary: 7,954 words
- Vocabulary density: 11.66%
- Named entity density: 7.30% (indicating a sparse labeling problem)

### 2. Entity Distribution Patterns
- **Class Imbalance**: Strong imbalance between entity types:
  - Indicators: 3.33% (largest entity class)
  - System: 1.94%
  - Malware: 1.32%
  - Organization: 0.59%
  - Vulnerability: 0.13% (smallest entity class)
- Non-entity tokens (O tag): 92.70%, indicating highly sparse labeling

### 3. Word Length Analysis
Interesting patterns in entity length characteristics:
- **Indicators** show highest variability and length (mean: 22.74, std: 19.62)
  - Maximum length of 113 characters suggests complex technical indicators
- **Vulnerabilities** (both B/I) have consistent longer lengths (~8.5)
  - Suggests formal vulnerability identifiers (e.g., CVE numbers)
- **Organization/System** entities show more compact lengths (5-6 characters)
  - Indicates shorter, standardized naming patterns

### 4. Entity Sequence Patterns
Notable transition patterns:
1. High I-Indicator → I-Indicator transitions (982)
   - Suggests indicators often contain multiple tokens
2. Regular entity start patterns:
   - O → B-System (803 occurrences)
   - O → B-Malware (689 occurrences)
3. Multi-token system entities:
   - B-System → I-System (314 occurrences)
   - Indicates complex system naming patterns

### 5. Statistical Significance
- Chi-square test (χ² = 3386.85, p < 0.001) confirms highly non-uniform distribution
- Indicates strong, non-random patterns in entity occurrences

## Implications for Classification
1. **Class Imbalance Handling**:
   - Need for specialized techniques to handle 92.70% non-entity tokens
   - Particular attention needed for rare classes (e.g., Vulnerabilities at 0.13%)

2. **Entity Length Considerations**:
   - Variable-length handling crucial for Indicators (1-113 characters)
   - Entity-specific length patterns could be valuable features

3. **Sequence Modeling**:
   - Strong sequential patterns suggest value in sequence modeling approaches
   - Entity-specific transition patterns could improve classification accuracy


````python
import pandas as pd
import numpy as np
import spacy
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict

class BTagAnalyzer:
    def __init__(self, nlp=None):
        self.nlp = nlp if nlp else spacy.load('en_core_web_sm')
        self.b_tag_contexts = defaultdict(list)
        self.pos_patterns = defaultdict(lambda: defaultdict(int))
        self.prev_words = defaultdict(lambda: defaultdict(int))
        self.positions = defaultdict(list)
        
    def process_file(self, file_path):
        """Process the file and extract B-tag patterns"""
        # Read file into a list of (word, tag) pairs
        sentences = []
        current_sentence = []
        
        with open(file_path, 'r') as f:
            for line in f:
                if line.strip():
                    word, tag = line.strip().split('\t')
                    current_sentence.append((word, tag))
                else:
                    if current_sentence:
                        sentences.append(current_sentence)
                        current_sentence = []
            if current_sentence:
                sentences.append(current_sentence)
        
        # Process each sentence
        for sentence in sentences:
            self._process_sentence(sentence)
                
    def _process_sentence(self, sentence):
        """Process a single sentence for B-tag patterns"""
        for i, (word, tag) in enumerate(sentence):
            if tag.startswith('B-'):
                entity_type = tag[2:]  # Remove B- prefix
                
                # Store position in sentence
                self.positions[entity_type].append(i / len(sentence))
                
                # Get previous word context if available
                if i > 0:
                    prev_word = sentence[i-1][0].lower()
                    self.prev_words[entity_type][prev_word] += 1
                    
                    # Get POS of previous word
                    doc = self.nlp(prev_word)
                    pos = doc[0].pos_
                    self.pos_patterns[entity_type][pos] += 1
                
                # Store full context
                context = {
                    'word': word,
                    'prev_word': sentence[i-1][0] if i > 0 else None,
                    'next_word': sentence[i+1][0] if i < len(sentence)-1 else None,
                    'position': i / len(sentence)
                }
                self.b_tag_contexts[entity_type].append(context)
    
    def plot_position_distributions(self):
        """Plot the distribution of B-tag positions in sentences"""
        plt.figure(figsize=(12, 6))
        for entity_type, positions in self.positions.items():
            sns.kdeplot(positions, label=entity_type)
        plt.title('Distribution of Entity Positions in Sentences')
        plt.xlabel('Relative Position (0=start, 1=end)')
        plt.ylabel('Density')
        plt.legend()
        return plt

    def plot_pos_patterns(self):
        """Plot heatmap of POS patterns before B-tags"""
        # Convert to DataFrame for visualization
        pos_df = pd.DataFrame({
            entity: dict(patterns)
            for entity, patterns in self.pos_patterns.items()
        }).fillna(0)
        
        plt.figure(figsize=(12, 8))
        sns.heatmap(pos_df, annot=True, fmt='.0f', cmap='YlOrRd')
        plt.title('POS Patterns Before B-tags')
        plt.ylabel('POS Tag')
        plt.xlabel('Entity Type')
        return plt
    
    def get_common_triggers(self, top_n=10):
        """Get most common trigger words for each entity type"""
        triggers = {}
        for entity_type, words in self.prev_words.items():
            sorted_words = sorted(words.items(), key=lambda x: x[1], reverse=True)
            triggers[entity_type] = sorted_words[:top_n]
        return triggers
    
    def get_insights(self):
        """Generate insights from the analyzed patterns"""
        insights = {
            'total_entities': {
                entity_type: len(contexts)
                for entity_type, contexts in self.b_tag_contexts.items()
            },
            'avg_positions': {
                entity_type: np.mean(positions)
                for entity_type, positions in self.positions.items()
            },
            'common_triggers': self.get_common_triggers(5),
            'pos_patterns': {
                entity_type: dict(sorted(patterns.items(), key=lambda x: x[1], reverse=True)[:5])
                for entity_type, patterns in self.pos_patterns.items()
            }
        }
        return insights
````

````python
# Initialize the analyzer with spacy
import spacy
nlp = spacy.load('en_core_web_sm')
analyzer = BTagAnalyzer(nlp)

# Process your data file
analyzer.process_file("data/mitre/train.txt")

# Get insights
insights = analyzer.get_insights()

# Print findings
print("=== Entity Distribution ===")
for entity_type, count in insights['total_entities'].items():
    print(f"{entity_type:15} {count:5d} entities")

print("\n=== Average Position in Sentence ===")
for entity_type, pos in insights['avg_positions'].items():
    print(f"{entity_type:15} {pos:.3f}")

print("\n=== Top Trigger Words ===")
for entity_type, triggers in insights['common_triggers'].items():
    print(f"\n{entity_type}:")
    for word, count in triggers:
        print(f"  {word:15} {count:4d}")

# Generate and show plots
analyzer.plot_position_distributions()
plt.show()

analyzer.plot_pos_patterns()
plt.show()
````

````python
import pandas as pd

def create_ner_df(text: str) -> pd.DataFrame:
    """
    Process NER tagged text and return a DataFrame with words, tags, and full sentences
    """
    words = []
    tags = []
    
    # Split text into lines and process each line
    for line in text.strip().split('\n'):
        if line.strip():  # Skip empty lines
            word, tag = line.strip().split('\t')
            words.append(word)
            tags.append(tag)
    
    # Create basic DataFrame
    df = pd.DataFrame({
        'word': words,
        'tag': tags
    })
    
    # Add sentence IDs (using periods, question marks, exclamation points as markers)
    sentence_ids = []
    current_sentence = 0
    
    for word in df['word']:
        sentence_ids.append(current_sentence)
        if word in {'.', '?', '!'}:
            current_sentence += 1
    
    df['sentence_id'] = sentence_ids
    
    # Create full sentence column
    df['full_sentence'] = df.groupby('sentence_id')['word'].transform(lambda x: ' '.join(x))
    
    return df

# Create DataFrame
# Read from file
with open('data/mitre/train.txt', 'r') as f:
    text = f.read()

# Create DataFrame
df = create_ner_df(text)

# Display first few rows
print("\nFirst few rows of the DataFrame:")
print(df.head(30))

# Get some basic statistics
print("\nEntity counts:")
print(df['tag'].value_counts())


# Create DataFrames for each category
malware_df = df[df['tag'].str.contains('Malware')]
system_df = df[df['tag'].str.contains('System')]
organization_df = df[df['tag'].str.contains('Organization')]
indicator_df = df[df['tag'].str.contains('Indicator')]
vulnerability_df = df[df['tag'].str.contains('Vulnerability')]
````

````python
import pandas as pd
import numpy as np
from collections import defaultdict

def analyze_entity_contexts(df, entity_type):
    """Analyze context patterns for a given entity type"""
    # Get rows with the entity
    entity_df = df[df['tag'].str.contains(entity_type)]
    
    # Word frequency analysis
    word_freq = entity_df['word'].value_counts()
    
    # Get previous words (context)
    prev_words = []
    next_words = []
    
    for sent_id in entity_df['sentence_id'].unique():
        sent_df = df[df['sentence_id'] == sent_id]
        entity_indices = sent_df[sent_df['tag'].str.contains(entity_type)].index
        
        for idx in entity_indices:
            if idx > 0 and (idx-1) in sent_df.index:
                prev_words.append(sent_df.loc[idx-1, 'word'])
            if (idx+1) in sent_df.index:
                next_words.append(sent_df.loc[idx+1, 'word'])
    
    prev_word_freq = pd.Series(prev_words).value_counts()
    next_word_freq = pd.Series(next_words).value_counts()
    
    return {
        'word_freq': word_freq,
        'prev_words': prev_word_freq,
        'next_words': next_word_freq
    }

# Analyze both entity types
system_analysis = analyze_entity_contexts(df, 'System')
org_analysis = analyze_entity_contexts(df, 'Organization')

print("=== Most Common System Words ===")
print(system_analysis['word_freq'].head(10))
print("\n=== Most Common Organization Words ===")
print(org_analysis['word_freq'].head(10))

print("\n=== Common Words BEFORE System Entities ===")
print(system_analysis['prev_words'].head(10))
print("\n=== Common Words BEFORE Organization Entities ===")
print(org_analysis['prev_words'].head(10))

print("\n=== Common Words AFTER System Entities ===")
print(system_analysis['next_words'].head(10))
print("\n=== Common Words AFTER Organization Entities ===")
print(org_analysis['next_words'].head(10))

# Find overlapping words
system_words = set(system_analysis['word_freq'].index[:20])
org_words = set(org_analysis['word_freq'].index[:20])
overlapping = system_words.intersection(org_words)

print("\n=== Overlapping Common Words ===")
print(list(overlapping))
````

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def create_feature_vector(df, entity_row_idx):
    """Create a feature vector for an entity occurrence"""
    row = df.loc[entity_row_idx]
    sent_df = df[df['sentence_id'] == row['sentence_id']]
    sent_indices = sent_df.index.tolist()
    current_pos = sent_indices.index(entity_row_idx)
    
    # Feature 1: Position in sentence (normalized)
    rel_position = current_pos / len(sent_indices)
    
    # Feature 2: Following technical word?
    technical_words = {'version', 'app', 'device', 'platform', 'store', 'software', 'service'}
    has_tech_follow = 0
    if current_pos < len(sent_indices) - 1:
        next_word = sent_df.iloc[current_pos + 1]['word'].lower()
        has_tech_follow = 1 if next_word in technical_words else 0
    
    # Feature 3: Preceded by business word?
    business_words = {'by', 'from', 'says', 'reported', 'according'}
    has_biz_before = 0
    if current_pos > 0:
        prev_word = sent_df.iloc[current_pos - 1]['word'].lower()
        has_biz_before = 1 if prev_word in business_words else 0
    
    # Feature 4: Contains digits?
    has_digits = 1 if any(c.isdigit() for c in row['word']) else 0
    
    # Feature 5: Word length
    word_length = len(row['word'])
    
    return [rel_position, has_tech_follow, has_biz_before, has_digits, word_length]

def analyze_entities(df):
    """Analyze and visualize entity patterns"""
    # Get indices for both entity types
    system_indices = df[df['tag'].str.contains('System')].index
    org_indices = df[df['tag'].str.contains('Organization')].index
    
    # Create feature arrays
    system_features = np.array([create_feature_vector(df, idx) for idx in system_indices])
    org_features = np.array([create_feature_vector(df, idx) for idx in org_indices])
    
    # Create scatter plot using two most distinguishing features
    plt.figure(figsize=(12, 8))
    
    # Plot relative position vs word length
    plt.scatter(system_features[:, 0], system_features[:, 4], 
               alpha=0.5, label='System', c='blue')
    plt.scatter(org_features[:, 0], org_features[:, 4], 
               alpha=0.5, label='Organization', c='red')
    
    plt.xlabel('Relative Position in Sentence')
    plt.ylabel('Word Length')
    plt.title('System vs Organization Entities - Feature Distribution')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # Add some statistics annotations
    plt.text(0.02, 0.95, f'System entities: {len(system_indices)}', 
             transform=plt.gca().transAxes)
    plt.text(0.02, 0.90, f'Organization entities: {len(org_indices)}', 
             transform=plt.gca().transAxes)
    
    # Print context patterns
    print("\nContext Analysis:")
    print("\nSystem Entities:")
    system_df = df.loc[system_indices]
    print("Most common words:", system_df['word'].value_counts().head(5).to_dict())
    print("Average position:", system_features[:, 0].mean())
    print("Average word length:", system_features[:, 4].mean())
    print("Technical word follows:", system_features[:, 1].mean())
    print("Business word precedes:", system_features[:, 2].mean())
    
    print("\nOrganization Entities:")
    org_df = df.loc[org_indices]
    print("Most common words:", org_df['word'].value_counts().head(5).to_dict())
    print("Average position:", org_features[:, 0].mean())
    print("Average word length:", org_features[:, 4].mean())
    print("Technical word follows:", org_features[:, 1].mean())
    print("Business word precedes:", org_features[:, 2].mean())
    
    plt.tight_layout()
    plt.show()

# Run the analysis
analyze_entities(df)
````
